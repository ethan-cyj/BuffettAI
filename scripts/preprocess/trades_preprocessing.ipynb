```python
import pandas as pd
```

```python
df = pd.DataFrame()

for i in range(14):
    temp = pd.read_excel(f"./brka_trades/BRKa Investor portfolio report ({i}).xlsx", skiprows=4)
    temp = temp.fillna(0)
    temp = temp.drop(columns=['Investor portfolio report'])

    temp_weightage = temp.iloc[:, 0:27] # Select weightage columns
    temp_weightage.columns = temp_weightage.iloc[1] # Set column names
    temp_weightage = temp_weightage.drop(columns=['Most Recent']) # Drop repeat column
    temp_weightage = temp_weightage.iloc[2:, :] # Drop first two rows
    temp_weightage = temp_weightage.melt(id_vars=['RIC', 'Security Name'], var_name='Date', value_name='Weightage') # Melt the dataframe

    temp_value = pd.concat([temp.iloc[:, 0:2], temp.iloc[:, 27:52]], axis=1)  # Select first 2 columns and value columns
    temp_value.columns = temp_value.iloc[1] # Set column names
    temp_value = temp_value.drop(columns=['Most Recent']) # Drop repeat column
    temp_value = temp_value.iloc[2:, :] # Drop first two rows
    temp_value = temp_value.melt(id_vars=['RIC', 'Security Name'], var_name='Date', value_name='Value') # Melt the dataframe

    temp_value_change = pd.concat([temp.iloc[:, 0:2], temp.iloc[:, 52:77]], axis=1)  # Select first 2 columns and value columns
    temp_value_change.columns = temp_value_change.iloc[1] # Set column names
    temp_value_change = temp_value_change.drop(columns=['Most Recent']) # Drop repeat column
    temp_value_change = temp_value_change.iloc[2:, :] # Drop first two rows
    temp_value_change = temp_value_change.melt(id_vars=['RIC', 'Security Name'], var_name='Date', value_name='Value Change') # Melt the dataframe

    temp_position = pd.concat([temp.iloc[:, 0:2], temp.iloc[:, 77:102]], axis=1)  # Select first 2 columns and value columns
    temp_position.columns = temp_position.iloc[1] # Set column names
    temp_position = temp_position.drop(columns=['Most Recent']) # Drop repeat column
    temp_position = temp_position.iloc[2:, :] # Drop first two rows
    temp_position = temp_position.melt(id_vars=['RIC', 'Security Name'], var_name='Date', value_name='Position') # Melt the dataframe

    temp_position_change = pd.concat([temp.iloc[:, 0:2], temp.iloc[:, 102:127]], axis=1)  # Select first 2 columns and value columns
    temp_position_change.columns = temp_position_change.iloc[1] # Set column names
    temp_position_change = temp_position_change.drop(columns=['Most Recent']) # Drop repeat column
    temp_position_change = temp_position_change.iloc[2:, :] # Drop first two rows
    temp_position_change = temp_position_change.melt(id_vars=['RIC', 'Security Name'], var_name='Date', value_name='Position Change') # Melt the dataframe

    temp_final = pd.merge(temp_weightage, temp_value, on=['RIC', 'Security Name', 'Date'])
    temp_final = temp_final.merge(temp_value_change, on=['RIC', 'Security Name', 'Date'])
    temp_final = temp_final.merge(temp_position, on=['RIC', 'Security Name', 'Date'])
    temp_final = temp_final.merge(temp_position_change, on=['RIC', 'Security Name', 'Date'])
    
    df = pd.concat([df, temp_final], axis=0)

# Drop if all values are 0
df = df[(df['Weightage'] != 0) & (df['Value'] != 0) & (df['Value Change'] != 0) & (df['Position'] != 0) & (df['Position Change'] != 0)].reset_index(drop=True)
df.to_excel('../../data/brka_trades.xlsx', index=False)
df
```
