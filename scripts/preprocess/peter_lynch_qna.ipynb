```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

PETER LYNCH Q&A TRANSCRIPT

```python
# Fetch the page content
url = "https://www.pbs.org/wgbh/pages/frontline/shows/betting/pros/lynch.html"
response = requests.get(url)

# Parse the HTML content
x = response.text
qn_tag = '<img align="left" alt="Q" src="../art/q.gif"/>'
qn_end = '</b></p><p>'
ans_tag = '<img align="left" alt="A" src="../art/a.gif"/></b>'
ans_end = '</p><p>'
# Regular expressions to extract the question and answer
question_pattern = re.escape(qn_tag) + r'(.*?)' + re.escape(qn_end)
answer_pattern = re.escape(ans_tag) + r'(.*?)' + re.escape(ans_end)

# Extract questions and answers using regex
questions = re.findall(question_pattern, x, re.DOTALL)
answers = re.findall(answer_pattern, x, re.DOTALL)

# Clean the extracted questions and answers
questions = [q.strip() for q in questions]
answers = [a.strip() for a in answers]

answers = [answer.strip() for answer in answers if answer.strip() != "You made ten times your money.  Is a ten bagger."]

# Create a DataFrame with questions and answers
peter_lynch_df = pd.DataFrame({
    'Question': questions,
    'Answer': answers,
    'Buffett_statement': 0
})
peter_lynch_df.to_csv('peter_lynch_qna.csv')
peter_lynch_df.head()
```

JOEL GREENBLATT Q&A TRANSCRIPT

```python
# Fetch the page content
url = "https://ritholtz.com/2018/04/transcript-joel-greenblatt/"
response = requests.get(url)

response_text = response.text

start = "GREENBLATT: Thanks.</p>"
end = "<p>END</p>"

pattern = re.escape(start) + r"(.*?)" + re.escape(end)
matches = re.findall(pattern, response_text, re.DOTALL)

speech = matches[0].strip()
cleaned_text = re.sub(r'<.*?>', '', speech)

# Regular expression to split text by speaker
pattern = r"(RITHOLTZ:|GREENBLATT:)(.*?)((?=RITHOLTZ:)|(?=GREENBLATT:)|$)"
matches = re.findall(pattern, cleaned_text, re.DOTALL)

# Create a list of questions and answers
data = []
question = None

for match in matches:
    speaker, dialogue = match[0], match[1].strip()
    
    if speaker == "RITHOLTZ:":
        if question is not None:  # Add the previous question-answer pair
            data.append([question, answer])
        question = dialogue
    elif speaker == "GREENBLATT:":
        answer = dialogue

# Append the last question-answer pair
if question is not None:
    data.append([question, answer])

# Convert to a DataFrame
df = pd.DataFrame(data, columns=["Question", "Answer"])

df = df.iloc[:(len(df)-1)]
df['Buffett_statement'] = 0
# df.to_csv('joel_greenblatt_qna.csv')
# df
```
